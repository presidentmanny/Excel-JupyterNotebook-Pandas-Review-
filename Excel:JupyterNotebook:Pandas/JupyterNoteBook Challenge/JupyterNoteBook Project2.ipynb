```python
# Dependencies
import pandas as pd
```

```python
# load CSV
new_coders='2016-FCC-New-Coders-Survey-Data.csv'
```

```python
# Read with pandas
# low_memory to get rid of error about mixed data type in some colums 
new_coders = pd.read_csv(new_coders, encoding = 'iso-8859-1', low_memory=False)
new_coders.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111

reduced_coders_pd=new_coders.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
```

```python
# Change reading 0 and 1 to No and Yes, respectively

reduced_coders_pd= reduced_coders_pd.replace({0.0: "No", 1.0: "Yes"})
reduced_coders_pd.head()
```

```python
# Extract rows for only those who attended a bootcamp
attended_bootcamp = reduced_coders_pd.loc[reduced_coders_pd["AttendedBootcamp"] == "Yes"]
attended_bootcamp.count()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
reduced_coders_pd['BootcampName']=new_coders['BootcampName']
boot_camps= reduced_coders_pd.loc[reduced_coders_pd["AttendedBootcamp"] == "Yes"]
boot_camps=boot_camps.BootcampName.value_counts()
boot_camps=pd.DataFrame(boot_camps)
boot_camps.reset_index(drop=False,inplace=True)
boot_camps.columns=['BootcampName','Count']
print(boot_camps.shape)
boot_camps.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recommend =reduced_coders_pd.loc[(reduced_coders_pd["AttendedBootcamp"] == "Yes")&(reduced_coders_pd["BootcampRecommend"] == "Yes")]
recommend=recommend.groupby('BootcampName').BootcampRecommend.count()

recommend=pd.DataFrame(recommend)
recommend.reset_index(drop=False,inplace=True)
print(recommend.shape)
recommend.head()
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merged_df=pd.merge(boot_camps,recommend,on='BootcampName')
merged_df.head()
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
merged_df['%Recommend']=100*merged_df['BootcampRecommend']/merged_df['Count']
```

```python
#Sort the new DataFrame in descending order of the percentage of recommenders you just calculated.
merged_df.sort_values('%Recommend',ascending=False,inplace=True)
```

```python
merged_df['%Recommend'] =merged_df['%Recommend'].map("{0:,.2f}%".format)
merged_df.head()
```

```python
# Export to excel and remove index
merged_df.to_excel("Percent_Recommend.xlsx", index=False)
```
